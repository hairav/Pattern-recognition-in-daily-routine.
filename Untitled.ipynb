```python
import pandas
import numpy
import pickle

FILENAME = 'data.csv'
HMM_DATA_FILE = 'HMM.obj'
DIC = {'b':0,'k':1,'w':2,'l':3}
DI = {0:'b',1:'k',2:'w',3:'l'}
```

```python
data = pandas.read_csv(FILENAME)
```

```python
print(data)
```

```python
HMM = []
day=[[], []]
for i in range(len(data['Room'])):
    day[0] = day[0] + [j for j in range(data['Start'][i],data['End'][i])]
    day[1] = day[1] + [DIC[data['Room'][i]] for j in range(data['Start'][i],data['End'][i])]
    if data['End'][i] == 1440:
        HMM.append([day[0][:1440],day[1][:1440]])
        day = [[],[]]
file = open(HMM_DATA_FILE, 'wb')
pickle.dump(HMM, file)
```

```python
Train = HMM[:120]
Test = HMM[120:]
Emmision ={}
for i in Train:
    for j in range(len(i[0])):
        if i[0][j] in Emmision:
            if i[1][j] in Emmision[i[0][j]]:
                Emmision[i[0][j]][i[1][j]] += 1
                continue
            Emmision[i[0][j]][i[1][j]] = 1
            continue
        Emmision[i[0][j]] = {i[1][j]:1}
#print(Emmision)
```

```python
for i in Emmision:
    s = 0
    for j in Emmision[i]:
        s += Emmision[i][j]
    for j in Emmision[i]:
        Emmision[i][j] /= float(s)
print(Emmision[0])
```

```python
trans = {}
trans['Start'] = {}
isStart = True
print(len(data['Room']))
for i in range(len(data['Room'])):
    if data['End'][i] == 1440:
        isStart =True
        continue
    if isStart:
        isStart = False
        if data['Room'][i] in trans['Start']:
             trans['Start'][data['Room'][i]] +=1
        else:
            trans['Start'][data['Room'][i]] = 1
    if data['Room'][i] in trans:
        if data['Room'][i+1] in trans[data['Room'][i]]:
            trans[data['Room'][i]][data['Room'][i+1]] += 1
            continue
        trans[data['Room'][i]][data['Room'][i+1]] = 1
        continue
    trans[data['Room'][i]] = {data['Room'][i+1] : 1}
print(trans)
```

```python
for i in trans:
    s=0
    for j in trans[i]:
        s += trans[i][j]
    for j in trans[i]:
        trans[i][j] /= s
print(trans)
```

```python
vtb = [[0, 0, 0, 0] for i in range(1440)]
btrack = [[0, 0, 0, 0] for i in range(1440)]
for i in range(1440):
    if i == 0:
        for j in DIC:
            try:
                vtb[i][DIC[j]] = Emmision[i][DIC[j]] * trans['Start'][j]
            except:
                continue
        continue
    for j in DIC:
        for k in DIC:
            try:
                if Emmision[i][DIC[j]] * trans[k][j] > vtb[i][DIC[j]]:
                    vtb[i][DIC[j]] = Emmision[i][DIC[j]] * trans[k][j]
                    btrack[i][DIC[j]] = k
            except:
                continue
print(btrack)
```

```python
labels = [0 for i in range(1440)]

for i in range(1439,0,-1):
    if i == 1439:
        m = max(vtb[i])
        labels[i] = vtb[i].index(m)
    labels[i-1] = DIC[btrack[i][labels[i]]]
p = labels[0]
for i in range(1439):
    if p != labels[i]:
        print(DI[labels[i]],i)
```
